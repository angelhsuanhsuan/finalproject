```python
from PIL import Image
from PIL import ImageTk
from PIL import ImageFont
from PIL import ImageDraw
from PIL import ImageEnhance
from tkinter import filedialog ,dialog
import cv2
import PIL.ImageOps
import numpy as np
import tkinter as tk
import tkinter.colorchooser as cc
```

```python
def resize_img(img ,size):
    if img.width > img.height:
        img_resize = img.resize((size ,int(img.height/img.width*size)))
    elif img.width < img.height:
        img_resize = img.resize((int(img.height/img.width*size) ,size))
    else:
        img_resize = img.resize((size ,size))
    return img.resize((size ,size))


def load_img_button():
    thumbnail_size = 200
    global img_tk ,file_path ,img ,img_filter
    
    load_text.delete("1.0","end")
    file_path = filedialog.askopenfilename(parent=win ,title="選擇要製作的圖片" ,initialdir="./image" ,filetypes=(("JPG files" ,"*.jpg") ,("ALL files" ,"*.*")))
    load_text.insert("insert" ,file_path)
    
    img = Image.open(file_path)
    
    img_filter = None
    img_resize = resize_img(img ,thumbnail_size)
    img_tk = ImageTk.PhotoImage(img_resize)
    img_label = tk.Label(width=200 ,height=200 ,bg="gray94" ,fg="blue" ,image=img_tk)
    img_label.place(x=50,y=100)

    
def save_img_button():
    global img_out
    
    file1_path = filedialog.asksaveasfilename(title="儲存圖片" ,initialdir="./image" ,filetypes=(('JPG files' ,'*.jpg') ,('ALL files' ,'*.*')))
    img_out.save(str(file1_path) + ".jpg")
    
```

```python
def word_input_button():
    global img_tk_out ,file_path ,img ,img_resize_out ,bk_img ,draw_img_out,img_out
    some_word1 = word1_text.get("1.0","end").strip()
    word1_size = word1_size_entry.get()
    some_word2 = word2_text.get("1.0","end").strip()
    word2_size = word2_size_entry.get()
    
    fontpath = "font/simsun.ttc"
    font1 = ImageFont.truetype(fontpath, size = int(word1_size))
    font2 = ImageFont.truetype(fontpath, size = int(word2_size))
    
    bk_img1 = cv2.imread(file_path)
    bk_img2 = cv2.cvtColor(bk_img1 ,cv2.COLOR_BGR2RGB)

    img_pil = Image.fromarray(bk_img2)
    img_array = np.asarray(img_pil)
    
    draw = ImageDraw.Draw(img_pil)
    #加入文字
    color1 = color1Print['bg']
    color2 = color2Print['bg']
    
    draw.text((70, 70), some_word1, font = font1, fill =color1)
    draw.text((350, 350), some_word2, font = font2, fill =color2)
    
    bk_img = np.array(img_pil)
    
    img_out = Image.fromarray(bk_img)
    img_resize_out = resize_img(img_out ,200)
    img_tk_out = ImageTk.PhotoImage(img_resize_out)
    img_label_out = tk.Label(image=img_tk_out)
    img_label_out.place(x=350,y=100)
    
```

```python
#圖片中文字1的顏色
def color1Choice():
    color1=cc.askcolor()
    color1 = color1[1]
    print(str(color1))
    color1Print.config(bg=color1)
    
#圖片中文字2的顏色    
def color2Choice():
    color2=cc.askcolor()
    color2 = color2[1]
    print(str(color2))
    color2Print.config(bg=color2)
```

```python
def edge_detection():
    global img_edge_resize_out ,img_tk_out,img_out
    img = cv2.imread(file_path)
    
    #邊緣檢測，圖案界線為白色，其餘為黑色
    img = cv2.GaussianBlur(img,(3,3),0)
    canny = cv2.Canny(img, 50, 150)

    canny_array = np.array(canny)
    img_out = Image.fromarray(canny_array)
    
    #反轉圖像顏色，黑變白，白變黑
    inverted_image = PIL.ImageOps.invert(img_out)
    
    bk_img = np.array(inverted_image)
    
    img_out = Image.fromarray(bk_img)
    img_resize_out = resize_img(img_out ,200)
    img_tk_out = ImageTk.PhotoImage(img_resize_out)
    img_label_out = tk.Label(image=img_tk_out)
    img_label_out.place(x=350,y=100)
    
```

```python
#飽和度調整器
def color_adjust(self):
    global img_adjust, img_adjust_tk,img_adjust_resize,img_out
    
#     img = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    img_word_out = Image.fromarray(bk_img)
    
    value = slider1.get()
    im1 = Image.open(file_path)
    
    enhance_im1 = ImageEnhance.Color(img_word_out)
    im2 = enhance_im1.enhance(value/5)
    
    im3 = np.array(im2)
    
    img_out = Image.fromarray(im3)
    img_adjust_resize = resize_img(img_out, 200)
    img_adjust_tk = ImageTk.PhotoImage(img_adjust_resize)
    img_adjust_label =tk.Label(width=200,height=200,bg ='gray94',fg='blue',image = img_adjust_tk)
    img_adjust_label.place(x=350,y=100)


#對比度調整器
def contrast_adjust(self):
    global img_adjust, img_adjust_tk,img_adjust_resize,img_out
    
    img_word_out = Image.fromarray(bk_img)
    
    value = slider2.get()
    im1 = Image.open(file_path)
    
    enhance_im1 = ImageEnhance.Contrast(img_word_out)
    im2 = enhance_im1.enhance(value/5)
    
    im3 = np.array(im2)
    
    img_out = Image.fromarray(im3)
    img_adjust_resize = resize_img(img_out,200)
    img_adjust_tk = ImageTk.PhotoImage(img_adjust_resize)
    img_adjust_label =tk.Label(width=200,height=200,bg ='gray94',fg='blue',image = img_adjust_tk)
    img_adjust_label.place(x=350,y=100)
#     img_adjust_label.grid(row=1, column=1, padx=25, pady=0, sticky='nw')
    
    
#亮度調整器    
def brightness_adjust(self):
    global img_adjust, img_adjust_tk ,img_adjust_resize,img_out
   
    img_word_out = Image.fromarray(bk_img)

    value = slider3.get()
    im1 = Image.open(file_path)
    
    enhance_im1 = ImageEnhance.Brightness(img_word_out)
    im2 = enhance_im1.enhance(value/5)
    
    im3 = np.array(im2)
    
    img_out = Image.fromarray(im3)
    img_adjust_resize = resize_img(img_out,200)
    img_adjust_tk = ImageTk.PhotoImage(img_adjust_resize)
    img_adjust_label =tk.Label(width=200,height=200,bg ='gray94',fg='blue',image = img_adjust_tk)
    img_adjust_label.place(x=350,y=100)
    
```
